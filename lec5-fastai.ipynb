```python
from fastai.data.transforms import RandomSplitter
trn_split,val_split=RandomSplitter(seed=42)(df)
```

```python
trn_indep,val_indep = t_indep[trn_split],t_indep[val_split]
trn_dep,val_dep = t_dep[trn_split],t_dep[val_split]
len(trn_indep),len(val_indep)
```

```python
def update_coeffs(coeffs, lr):
    coeffs.sub_(coeffs.grad * lr)
    coeffs.grad.zero_()
```

```python
def one_epoch(coeffs, lr):
    loss = calc_loss(coeffs, trn_indep, trn_dep)
    loss.backward()
    with torch.no_grad(): update_coeffs(coeffs, lr)
    print(f"{loss:.3f}", end="; ")
```

```python
def init_coeffs(): 
    return (torch.rand(n_coeff)-0.5).requires_grad_()
```

```python

def train_model(epochs=30, lr=0.01):
    torch.manual_seed(442)
    coeffs = init_coeffs()
    for i in range(epochs): 
        one_epoch(coeffs, lr=lr)
    return coeffs
```

## In function
- our n_coeffs in excel is 2
- 0.5 is centerised - if gradients too big or too large its not going to point in right direction so training would not be possible
- So we divide by n_hidden
- I dont need to add bias for first layer
- But I need for second layer the constant term so const added
- Here single scaler random number
  

```python
def init_coeffs(n_hidden=20):
    layer1 = (torch.rand(n_coeff, n_hidden)-0.5)/n_hidden
    layer2 = torch.rand(n_hidden, 1)-0.3
    const = torch.rand(1)[0]
    return layer1.requires_grad_(),layer2.requires_grad_(),const.requires_grad_()
     
```

## calc preds
- coeffs is above function return its tuple
- matrix multiply
- F.relu matrix replace negative to positive Check function
- another matrix multiply

```python
import torch.nn.functional as F

def calc_preds(coeffs, indeps):
    l1,l2,const = coeffs
    res = F.relu(indeps@l1)
    res = res@l2 + const
    return torch.sigmoid(res)
```

# Update
- for each later we are updating loop because of 3 tuples we have in return
- apply same things

```python
def update_coeffs(coeffs, lr):
    for layer in coeffs:
        layer.sub_(layer.grad * lr)
        layer.grad.zero_()
```

## Error Fixed

```python
def calc_preds(coeffs, indeps):
    l1,l2,const = coeffs
    
    if indeps.dtype != torch.double:
        indeps = indeps.double()

    if l1.dtype != torch.double:
        l1 = l1.double()

    res = F.relu(indeps@l1)
    
    if res.dtype != torch.double:
        res = res.double()

    if l2.dtype != torch.double:
        l2 = l2.double()
    
    res = res@l2 + const
    return torch.sigmoid(res)
```

## Deep Learning
- 2 hidden layers 10 activation in each. you can add for each hidden as many as u want
- for example [10,10] 10 neurons in first hidden 10 neurons in second hiddne 
- first initialise for each hidden we torch.rand
- then in the we did rand for const
- then looping for l.require_grads
- Paperpen spread sheet u can use

```python
def init_coeffs():
    hiddens = [10, 10]  # <-- set this to the size of each hidden layer you want
    sizes = [n_coeff] + hiddens + [1]
    n = len(sizes)
    layers = [(torch.rand(sizes[i], sizes[i+1])-0.3)/sizes[i+1]*4 for i in range(n-1)]
    consts = [(torch.rand(1)[0]-0.5)*0.1 for i in range(n-1)]
    for l in layers+consts: l.requires_grad_()
    return layers,consts
```

## Common Mistake

## Score Rate Concept
- Good split is the one which all the values of dependent variable on one side are pretty much the same and on the other sides pretty much same
- for example see chart pretty much all the males had same survival outcome which " did not survive" and all the females had about same survival outcome which is "they did survive" that would be good split. that would be good split
- you want each of group to be similar as possible on dependent variable
- _side_score : tells how similar those groups are : standard deviations we basically added the standard deviation of two groups of dependent variable
- if its small std but its really small group not.  intersting for us.  multiply by size
- score: we can see in function score of left and right and take average
- score(trn[sex], trn_y,0.5) if we split based on 0.5 less or greater  create two groups gives 0.4 score
- LogFare we do for 2.7 gives 0.47 so 0.4 is better
- we have interactive tool dropdown we can choose variables we can do for cat and con

```python

def _side_score(side, y):
    tot = side.sum()
    if tot<=1: return 0
    return y[side].std()*tot
```

```python
   
def score(col, y, split):
    lhs = col<=split
    return (_side_score(lhs,y) + _side_score(~lhs,y))/len(y)
```

```python
def min_col(df, nm):
    col,y = df[nm],df[dep]
    unq = col.dropna().unique()
    scores = np.array([score(col, y, o) for o in unq if not np.isnan(o)])
    idx = scores.argmin()
    return unq[idx],scores[idx]

min_col(trn_df, "Age")
```
